```python
from Bio import Align
aligner = Align.PairwiseAligner()
alignments = aligner.align("TACCG", "ACG")
for alignment in sorted(alignments):
    print("Score = %.1f:" % alignment.score)
    print(alignment)
```

```python
aligner.mode = 'local'
alignments = aligner.align("TACCG", "ACG")
for alignment in sorted(alignments):
    print("Score = %.1f:" % alignment.score)
    print(alignment)
```

```python
# Smith - Waterman
aligner.mode = 'local'
print(aligner.algorithm)
alignments = aligner.align("TACCG", "ACG")
for alignment in sorted(alignments):
    print("Score = %.1f:" % alignment.score)
    print(alignment)
```

```python
# Needleman - Wunsch
aligner.mode = 'global'
print(aligner.algorithm)
aligner.match_score = 2
aligner.mismatch_score = -1
for alignment in aligner.align("TACCG", "ACG"):
    print("Score = %.1f:" % alignment.score)
    print(alignment)
```

```python
aligner.open_gap_score = -0.5
aligner.extend_gap_score = -0.1
aligner.target_end_gap_score = 0.0
aligner.query_end_gap_score = 0.0
for alignment in aligner.align("TACCG", "ACG"):
    print("Score = %.1f:" % alignment.score)
    print(alignment)
```

```python
from Bio.Align import substitution_matrices
aligner = Align.PairwiseAligner()
aligner.substitution_matrix = substitution_matrices.load("BLOSUM62")
alignments = aligner.align("KEVLA", "EVL")
alignments = list(alignments)
print("Number of alignments: %d" % len(alignments))
alignment = alignments[0]
print("Score = %.1f" % alignment.score)
print(alignment)
```

```python
from Bio import SeqIO
reference_record = SeqIO.read("../data/nc_045512.fasta", "fasta")
record_iterator = SeqIO.parse("../data/sc_covid_top_200.fasta", "fasta")
child_record = next(record_iterator)
from Bio import Align

aligner = Align.PairwiseAligner()
aligner.mode = 'global'
aligner.open_gap_score = -8
aligner.extend_gap_score = -8
# aligner.target_end_gap_score = 0.0
# aligner.query_end_gap_score = 0.0
aligner.match_score = 2
aligner.mismatch_score = -1
alignments = aligner.align(reference_record.seq,child_record.seq)
```

```python
alignment = next(alignments)
```

```python
print("Score = %.1f" % alignment.score)
print(alignment)
```

```python
# show with terminal as well 
ref = SeqIO.read('../data/NC_045512.fna','fasta')
rest = SeqIO.parse('../data/sc_covid_top_200.fasta','fasta')
with open('unaligned.fasta','w') as f:
    SeqIO.write(ref,f,'fasta')
    SeqIO.write(rest,f,'fasta')
```
