```python
best_hit = df.sort_values([1,2,3],ascending=[False,False,False]).reset_index(drop=True).iloc[0]
name,score,perc_id,length,qstart,qend = best_hit.to_list()
```

```python
if(length == 19 and perc_id >= 94.5):
    print("+500")
else:
    print("-500")
```

```python
for contig in SeqIO.parse(f'../data/spades/contigs.fasta','fasta'):
    if(contig.name == name):
        print(len(contig),contig.name)
        extracted_contig = contig
```
